```python
import numpy as np
import pandas as pd
```

# Series form LIST

```python
# series from lists

country = ["Nepal", "India", "Bhutan", "USA", "Australia"]
pd.Series(country)
```

```python
# integers

runs = [10,32,10,101,65]
pd.Series(runs)
```

```python
# custom index

marks = [66, 70,64,68,72]
subs = ["nepali", "social", "english", "account", "maths"]

pd.Series(marks, index=subs)
```

```python
#  setting a name

marks = [66, 70,64,68,72]
subs = ["nepali", "social", "english", "account", "maths"]

my_marks = pd.Series(marks, index=subs, name = "jiban mark's")
my_marks
```

# Series from Dictionary

```python
marks = {
    "english" : 64,
    "nepali" : 66,
    "social" : 70,
    "maths" : 72,
    "computer" : 47,
    "account" : 70
}

marks_series = pd.Series(marks, name = "Jiban Marks")
print(marks_series)
```

# Series Attributes

```python
#  size

marks_series.size
```

```python
# name
marks_series.name
```
