```python
ipl = pd.read_csv("ipl-matches.csv")
ipl.head(3)
```

```python
#  find which player has won most of the potm -> in finals and qualifiers

print(ipl[ ~ ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts())
```

```python
# Toss Decision Plot
ipl['TossDecision'].value_counts().plot(kind = "pie")
```

```python
# how many matches each team has played
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending=False)
```
